<font size="6"> List, Vector Array, For Loop

```python
# Transform a list of values and plot it 

# Create an array
vec_x  = np.array([-5,-3,0,10,20,40])
print(vec_x)
```

```python
# Create a new vector vec_y with formula cos
vec_y=[]

index = 0
for x in vec_x:
    y =(np.cos(x))**(2)
    vec_y.append(y)
index = index + 1

print(vec_y)
```

<font size="6"> Scatterplot

```python
# Produce a scatter plot of "vec_y" against "vec_x"
# Label axes and title
#Change the color to "red"

plt.scatter( x= vec_x , y = vec_y, color = "red")
plt.xlabel("X Value")
plt.ylabel("Y Value")
plt.show()
```

```python

# Create two lists
list_fruits = ["apple", "orange", "pear", "banana"]
list_colors = ["blue", "red","green", "orange"]

# Create a new variable message = "orange"
message = "orange"

# Create an if/else that tests whether message is a fruit and a color. 
# Print a message if the test is true.

if message in list_fruits:
    print("message is a fruit")
else:
    print("message is not a fruit")

if message in list_colors:
    print("message is a color")
else:
    print("message is not a color")
```

<font size="6"> Plot Random Variables

```python
# Let n =1000
# Generate "vec_x" from a Normal with (loc = 2, scale = 1, size = n)
# Generate "vec_z" from a chi square with (df = 2, size = n)
# Generate "vex_y" from a uniform with (low = 5,high = 10, size = n)
# Plot two scatter plots: (vec_y vs vec_x) and (vec_z vs vec_x)
# Label the axes and title
# Each graph should have a different title
# To get full points, plot the histograms on a single row using the
# subplots syntax (see Lecture 6)
```

```python
n = 1000

vec_x  = np.random.normal(loc = 2, scale = 1, size = n)
vec_z  = np.random.chisquare(df = 2, size = n)
vec_y  = np.random.uniform(low = 5,high = 10, size = n)

fig, list_subfig = plt.subplots(1, 2, figsize = (6,3))
plt.tight_layout()

list_subfig[0].scatter(x = vec_x, y = vec_y)
list_subfig[0].set_title("Normal vs Uniform")
list_subfig[0].set_xlabel("Normal")
list_subfig[0].set_ylabel("Uniform")

list_subfig[1].scatter(x = vec_x, y = vec_z)
list_subfig[1].set_title("Normal vs Chisquare")
list_subfig[1].set_xlabel("Normal")
list_subfig[1].set_ylabel("Chisquare")
```
